## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns

y_df = attrition_df[["Attrition", "Department"]]
y_df
```

```python
# Create a list of at least 10 column names to use as X data
list_x = ["Education", "Age", "DistanceFromHome", "JobSatisfaction", "OverTime", "JobRole", "WorkLifeBalance", "YearsAtCompany", "NumCompaniesWorked", "YearsInCurrentRole"]


# Create X_df using your selected columns
X_df = attrition_df[list_x]

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=42)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary

encoder = OneHotEncoder(sparse_output=False)

X_train_enc = encoder.fit_transform(X_train["JobRole"].values.reshape(-1, 1))
X_test_enc = encoder.transform(X_test["JobRole"].values.reshape(-1, 1))

X_train_ohe = pd.DataFrame(X_train_enc, columns=encoder.get_feature_names_out())
X_test_ohe = pd.DataFrame(X_test_enc, columns=encoder.get_feature_names_out())

X_train_ohe.head()
```

```python

label = LabelEncoder()

X_train_lbl = label.fit_transform(X_train["OverTime"])
X_test_lbl = label.transform(X_test["OverTime"])

X_train_lbl_df = pd.DataFrame(X_train_lbl, columns=["OverTime"])
X_test_lbl_df = pd.DataFrame(X_test_lbl, columns=["OverTime"])

X_train_lbl_df.head()
```

```python
X_train = X_train.drop(["OverTime", "JobRole"], axis=1)

X_train.head()
```

```python
X_test = X_test.drop(["OverTime", "JobRole"], axis=1)

X_test.head()
```

```python
# Create a StandardScaler
scaler = StandardScaler()


# Fit the StandardScaler to the training data
scaler.fit(X_train)

# Scale the training and testing data
scaler.transform(X_train)
scaler.transform(X_test)
```

```python
# Create a OneHotEncoder for the Department column
encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
y_train_ohe = encoder.fit_transform(y_train[["Department"]])
y_test_ohe = encoder.transform(y_test[["Department"]])

columns = encoder.get_feature_names_out(["Department"])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_dep_enc = pd.DataFrame(y_train_ohe, columns=columns)
y_test_dep_enc = pd.DataFrame(y_test_ohe, columns=columns)

```

```python
# Create a OneHotEncoder for the Attrition column
encoder_1 = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
y_train_att_ohe = encoder_1.fit_transform(y_train[["Attrition"]])
y_test_att_ohe = encoder_1.transform(y_test[["Attrition"]])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_att_enc = pd.DataFrame(y_train_att_ohe, columns=encoder_1.get_feature_names_out())
y_test_att_enc = pd.DataFrame(y_test_att_ohe, )


```

```python
X_train.head()
```

```python
print(f"X_train shape: {X_train.shape}")
print(f"y_train_dep_enc shape: {y_train_dep_enc.shape}")
print(f"y_train_att_enc shape: {y_train_att_enc.shape}")
```

```python
X_train_enc = pd.concat([X_train, X_train_lbl_df, X_train_ohe], axis=1)
X_test_enc = pd.concat([X_test, X_test_lbl_df, X_test_ohe], axis=1)
```

```python
X_train_enc = X_train_enc[:1102]
X_test_enc = X_test_enc[:368]
display(X_test_enc.shape)
display(X_train_enc.shape)
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
len(X_train_enc)

# Create the input layer
input_layer = layers.Input(shape=(18, ), name="input_layer")

# Create at least two shared layers
shared_layer1 = layers.Dense(64, activation='relu')(input_layer)
shared_layer2 = layers.Dense(128, activation='relu')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
hidden_layer_department = layers.Dense(32, activation='relu')(shared_layer2)

# Create the output layer
department_output = layers.Dense(3, activation='softmax', name='department_output')(hidden_layer_department)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
hidden_layer_attrition = layers.Dense(32, activation='relu')(shared_layer2)

# Create the output layer
attrition_output = layers.Dense(2, activation='sigmoid', name='attrition_output')(hidden_layer_attrition)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[department_output, attrition_output])

# Compile the model
model.compile(loss={"department_output": "categorical_crossentropy",
              "attrition_output": "binary_crossentropy"},
              optimizer="adam",
              metrics={"department_output": "accuracy",
                       "attrition_output": "accuracy"})

# Summarize the model
model.summary()
```

```python
print(f"X_train_enc shape: {X_train_enc.shape}")
print(f"y_train_dep_enc shape: {y_train_dep_enc.shape}")
print(f"y_train_att_enc shape: {y_train_att_enc.shape}")
```

```python
# Train the model

train_model = model.fit(X_train_enc,
                        {"department_output": y_train_dep_enc,
                         "attrition_output": y_train_att_enc},
                        epochs=75,
                        batch_size=32,
                        validation_split=0.2,
                        verbose=1)
```

```python
display(X_test_enc.shape)
display(y_test_att_enc.shape)
```

```python
# Evaluate the model with the testing data
model_results = model.evaluate(X_test_enc,
                               {"department_output": y_test_dep_enc,
                                "attrition_output": y_test_att_enc},
                               verbose=1)
model_results
```

```python
# Print the accuracy for both department and attrition
department_accuracy = model_results[1]
attrition_accuracy = model_results[2]

print(f"Department predictions accuracy: {department_accuracy}")
print(f"Attrition predictions accuracy: {attrition_accuracy}")
```
